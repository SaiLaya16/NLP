**TASK 2**

```python
import re

def preprocess_text(text):
    # 1. lowercase
    text = text.lower()
    
    # 2. remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 3. remove emails
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. remove emojis / non-ascii characters
    text = text.encode('ascii', 'ignore').decode()
    
    # 6. normalize repeated characters (soooo -> so, goooood -> good)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 7. remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # 8. remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 9. remove short words (≤2) except "no", "not"
    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]
    
    return " ".join(words)


# Example test
text = "WOW!!! This is sooo goooood!!! Visit http://example.com now 123"
print(preprocess_text(text))
```

**TASK 3**

```python
texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in texts:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()

    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

**TASK 4**

```python
def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total if total > 0 else 0
    return total, unique, round(avg_len,2)

for text in texts:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()

    total, unique, avg = token_analysis(tokens)

    print("Text:", text)
    print("Tokens:", tokens)
    print("Total:", total, "| Unique:", unique, "| Avg Length:", avg)
    print("-" * 50)
```

**TASK 5**

```python
from collections import Counter

all_tokens = []

for text in texts:
    cleaned = preprocess_text(text)
    tokens = cleaned.split()
    all_tokens.extend(tokens)

freq = Counter(all_tokens)
top_10 = freq.most_common(10)
least_common = [word for word, count in freq.items() if count == 1][:5]
print("Top 10 Most Frequent Words:")
for word,count in top_10:
    print(word,":",count)
print("\nTop 5 Least Frequent Words:")
for word in least_common:
    print(word, ":", freq[word])
```

**TASK 6**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        cleaned = preprocess_text(text)
        tokens = cleaned.split()
        clean_sentences.append(cleaned)
        all_tokens.extend(tokens)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
texts = ["", "😍😍😍", "12345", "Hello World!!!"]
print(full_pipeline(texts))
```

```python
def preprocess_text(text):
    import re
    
    if not text or text.strip() == "":
        return ""

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)

    text = text.encode('ascii', 'ignore').decode()

    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]

    return " ".join(words)

tests = ["", "😍😍😍", "12345"]

for t in tests:
    print("Input:", t)
    print("Output:", preprocess_text(t))
    print("-" * 30)
```
