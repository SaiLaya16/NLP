```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# ✅ Load model
model_name = "google/flan-t5-base"

print("Loading model... ⏳")

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
model.eval()

print("✅ Model loaded successfully!")
print("===================================")
print("🤖 Chatbot: Hello! I am your AI assistant.")
print("🤖 Type 'exit' or 'quit' to stop chatting.")
print("===================================\n")

# ✅ Chat loop
while True:
    try:
        user_input = input("You: ").strip().lower()

        # Exit condition
        if user_input in ["exit", "quit"]:
            print("🤖 Chatbot: Goodbye! Have a great day 😊")
            break

        # Empty input
        if user_input == "":
            print("🤖 Chatbot: Please enter something!")
            continue

        print("🤖 Thinking...")

        # ✅ Handle basic conversation
        if user_input in ["hi", "hello", "hey"]:
            print("🤖 Chatbot: Hello! How can I help you today?")
            continue

        if "how are you" in user_input:
            print("🤖 Chatbot: I'm doing well! How can I assist you?")
            continue

        if "your name" in user_input:
            print("🤖 Chatbot: I am an AI chatbot created using Hugging Face Transformers.")
            continue
        
        if "python" in user_input:
            print("🤖 Chatbot: Python is used for web development, data analysis, machine learning, and automation.")
            continue
        
        if "sky blue" in user_input:
            print("🤖 Chatbot: The sky appears blue because blue light scatters more in the atmosphere due to its shorter wavelength.")
            continue
        
        if "artificial intelligence" in user_input:
            print("🤖 Chatbot: Artificial Intelligence is the simulation of human intelligence in machines.")
            continue
        
        if "who created python" in user_input:
            print("🤖 Chatbot: Python was created by Guido van Rossum in 1991.")
            continue


        prompt = f"""
You are an intelligent AI assistant.
Give a correct, detailed and meaningful answer in simple English.

If the question is factual, give the correct fact.
If the question is conceptual, explain clearly.

Question: {user_input}
Answer:
"""

        inputs = tokenizer(prompt, return_tensors="pt")

        outputs = model.generate(
            **inputs,
            max_new_tokens=120,
            do_sample=False,
            num_beams=5,
            early_stopping=True
        )

        response = tokenizer.decode(outputs[0], skip_special_tokens=True)
        if len(response.strip()) < 5 or "AI assistant" in response.lower():
            response = "I'm not sure, but I can help explain that better. Please ask again more clearly."

        print("🤖 Chatbot:", response)

    except Exception as e:
        print("⚠️ Error occurred:", str(e))
        print("Please try again.\n")
```
